```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression
```

```python
#path = "student_scores - student_scores.csv", data = pd.read_csv(path)
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("The data is imported sucessfully",)
data.head()
```

Let's plot our data points on 2-D graph to eyeball our dataset and see if we can manually find any relationship between the data. We can create the plot with the following script:

```python
data.plot(x='Hours', y='Scores', style='o', markeredgecolor='yellow', markerfacecolor='Red' )
plt.title('Study Hours vs Percentage Scored')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Scored')
plt.grid(True, color='blue', linestyle= "-.")
plt.show()
```

### **Data Manipulation_Dividing the data into train & test sets**
We split our data into training and testing sets, with 80:20 ratio 

```python
# Split Data for trainig:

from sklearn.model_selection import train_test_split  


x = data.loc[:,"Hours"].values  
y = data.loc[:,"Scores"].values  
x_train , x_test , y_train , y_test = train_test_split(x , y,test_size=0.2, random_state=0) 

# We Do this because there is just one feature, and the mode expecting 2D array:
x_train = x_train.reshape(-1,1) 
x_test = x_test.reshape(-1,1) 

print("Train Data:")
print("x : {}".format(x_train.shape))
print("y : {}".format(y_train.shape))
print("-" * 79)
print("Test Data:")
print("x : {}".format(x_test.shape))
print("y : {}".format(y_test.shape))
print("-" * 79)
print("Our data is ready for training !! ")
```

### **Training the Algorithm**
We have split our data into training and testing sets, and now is finally the time to train our algorithm. 

```python
# Trainig time ^-^ .. We will use LinearRegression because the data is simple (2 Features Only).

from sklearn.linear_model import LinearRegression  
regression = LinearRegression()  
regression.fit(x_train, y_train) 
print("Trainig Complete !")
```

### **Making Predictions**
Now that we have trained our algorithm, it's time to make some predictions.

```python
print(x_test) # Testing data - In Hours
y_pred = regression.predict(x_test) # Predicting the scores

# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```
